```python
rango=linspace(20,20,500) #Si c!=0 tenemos que ajustar este rango
def plot_Taylor (p,c,n): #La función solo funciona con polinomios
    d=[0]*(n+1)
    rango=linspace(c-20,c+20,500)
    derivadas=[0]*(n+1)
    derivadas[0]=p
    polinomio_taylor=np.poly1d([0]*(n+1))
    polinomio_taylor[0]=p(c)
    for i in range(1,n+1):
        derivadas[i]=derivadas[i-1].deriv() #ponemos en una lista las derivadas
        polinomio_taylor[i]=derivadas[i](c)/factorial(i) #y en otra los coeficientes de Taylor
    plt.plot(rango,polinomio_taylor(rango-c)); #graficamos nuestra aproximación
    plt.plot(rango,p(rango))

p=np.poly1d([5,3,0,1,6])
n=10 #número de iteraciones de Taylor que queremos
c=5 #donde queremos centrar nuestra serie
for i in range(1,n):
   plot_Taylor(p,c,i)
```
